```python
#To create augmented MNIST Datasets by appyling rotational and center crop transformations
import numpy as np
import pandas as pd
import numpy
from scipy import ndimage
import cv2
from sklearn.datasets import fetch_openml
from sklearn.metrics.cluster import completeness_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from matplotlib import offsetbox
from sklearn import (manifold, datasets, decomposition)
import numpy as np
import time
from sklearn import cluster
from sklearn.neighbors import kneighbors_graph
from sklearn.utils import shuffle
from matplotlib import offsetbox
import pickle
# Fetching the dataset from 
# mldata.org/repository/data/viewslug/mnist-original through the sklearn helper
mnistt = fetch_openml('mnist_784')
data = mnistt.data.reshape((mnistt.data.shape[0], 28, 28))
data_cropp = np.zeros((70000, 14, 14)) #initializing MNIST-CROP augmented dataset
data_60 = data.copy()
data_90 = data.copy() #not used in Final Report
data_120 = data.copy()
data_150 = data.copy() #not used in Final Report
data_180 = data.copy()

def crop_center(img,cropx,cropy): #function to crop MNIST images from the center
    y,x = img.shape
    startx = x//2-(cropx//2)
    starty = y//2-(cropy//2)    
    return img[starty:starty+cropy,startx:startx+cropx]

for h in range(len(data)-1):
    data_cropp[h] = crop_center(data[h], 14, 14)
for i in range(len(data) - 1):
    data_60[i] = ndimage.rotate(data_60[i],60,reshape=False, cval=np.median(data[i]))
for j in range(len(data) -1):
    data_90[j] = ndimage.rotate(data_90[j],90,reshape=False, cval=np.median(data[i])) #not used in Final Report
for k in range(len(data) -1):
    data_120[k] = ndimage.rotate(data_120[k],120,reshape=False, cval=np.median(data[i]))
for l in range(len(data) -1):
    data_150[l] = ndimage.rotate(data_150[l],150,reshape=False, cval=np.median(data[i])) #not used in Final Report
for m in range(len(data) -1):
    data_180[m] = ndimage.rotate(data_180[m],180,reshape=False, cval=np.median(data[i]))

with open('dataaugment1.pickle', 'wb') as f: #storing variable for use in OTDGWD.ipynb
            pickle.dump([data_cropp, data_60, data_90, data_120, data_150, data_180], f)
    
    




```
